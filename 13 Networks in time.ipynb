```python
import networkx as nx
import matplotlib.pyplot as plt
import datetime
import time
```

```python
G_wiki = nx.read_edgelist('./data/ligtenberg2017/wikilinks.csv'
                          , data = [('begin', int), ('end', int)]
                          , create_using = nx.MultiGraph
                         )
```

```python
len(G_wiki.nodes)
```

## What is here?
- Link between article $a$ and $b$, and the begin and end date of this link.

## Our task

- Extract snapshots for visualize evolution.
- Quantify the evolution. 


```python
G_wiki.edges['10290', '56346', 0]
```

```python
def get_snapshot(G, date):
    # Convert dates into timestamp
    dt = datetime.datetime.strptime(date, '%Y-%m-%d')
    timestamp = time.mktime(dt.timetuple())
    
    # Find edges that existed in this timestamp
    snapshot_edges = []
    for e in G.edges:
        if G.edges[e]['begin'] <= timestamp and G.edges[e]['end']>= timestamp:
            snapshot_edges.append(e)
    
    return nx.Graph(G.edge_subgraph(snapshot_edges))   
    
```

```python
G_test = get_snapshot(G_wiki, '2010-10-01' )
nx.draw(G_test)
```

```python
# Visualize the evolution on certain dates
dates = ['2010-10-01', '2010-10-08', '2010-10-15', '2010-10-22', '2010-10-29', '2010-11-06']

# Create a figure for all subplots
plt.figure(figsize=(10,15))

pos = None
for i, date in enumerate(dates):
    # Get a snapshot of the network on the date
    G = get_snapshot(G_wiki, date)
    
    plt.subplot(3, 2, 6-i) # 3 rows, 2 columns, i = id of subplot
    plt.title(date)
    
    # Calculate layout for the network
    pos = nx.spring_layout(G, pos=pos)
    
    #Visualize
    nx.draw_networkx(G, pos=pos, node_size=0, with_labels=False)
```

### Clustering

Measure of concentration of edges around a node.

For a node $v$, we look at the neighbors, assume it has $k$. 

Maximum number of edges between $k$ vertices: $\frac{k(k-1)}{2}$.

Local clustering: edges between neighbors / maximum number of edges.

Average clustering: is the average of local clustering values on all vertices.

```python
# Evolution of clustering measure in the network
year = 2001
month = 10
clustering = []

for i in range(24):
    date = '{}-{}-01'.format(year, month)
    G = get_snapshot(G_wiki, date)
    clustering.append(nx.average_clustering(G))
    
    # Update month and year
    month += 1
    if month > 12:
        month -= 12
        year += 1 
```

```python
plt.plot(clustering)
```
